#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
df = pd.read_csv('vehicles.csv')
df.head()
```

```python
plt.scatter(df["Combined MPG"], df["Highway MPG"])
plt.title('Vehicles Scatter plot')
plt.xlabel('Combined MPG')
plt.ylabel('Highway MPG')
plt.show()
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# your code here

fig1, ax1 = plt.subplots()
ax1.set_title('CO2 Emission Grams/Mile')
ax1.boxplot(df["CO2 Emission Grams/Mile"], notch =True)
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
x=df["Fuel Barrels/Year"]
n_bins = 15
plt.hist(x,n_bins,facecolor='green')
plt.title("histogram of the Fuel Barrels/Year")
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
x=df["Fuel Type"]
y=df["City MPG"]
plt.bar(x,y)
plt.xticks("Fuel Type")
plt.ylabel('Usage')
plt.title('Fuel Type &City MPG')
plt.show
```
