# Challenge 1 - Who Have Published What At Where?

```python
import pandas as pd

df = pd.read_sql_query("""SELECT au.au_id as "AUTHOR ID", au.au_lname as "LAST NAME", au.au_fname as "FIRST NAME", t.title as"TITLE", pub.pub_id as "PUBLISHER"
 FROM authors AS au
INNER JOIN titleauthor as ta 
ON au.au_id=ta.au_id
INNER JOIN titles as t
ON ta.title_id=t.title_id
INNER JOIN publishers as pub
ON t.pub_id=pub.pub_id;""", engine)
df

```

# Challenge 2 - Who Have Published How Many At Where?

```python
df2 = pd.read_sql_query("""SELECT au.au_id as "AUTHOR ID", au.au_lname as "LAST NAME", au.au_fname as "FIRST NAME", t.title as"TITLE", pub.pub_id as "PUBLISHER",count(t.title ) as "TITLE COUNT"
 FROM authors AS au
INNER JOIN titleauthor as ta 
ON au.au_id=ta.au_id
INNER JOIN titles as t
ON ta.title_id=t.title_id
INNER JOIN publishers as pub
ON t.pub_id=pub.pub_id
GROUP BY au.au_id,au.au_lname,au.au_fname,t.title,pub.pub_id;""", engine)
df2
```

# Challenge 3 - Best Selling Authors

```python
df3 = pd.read_sql_query("""SELECT au.au_id as "AUTHOR ID", au.au_lname as "LAST NAME", au.au_fname as "FIRST NAME", sum(s.qty) as "TOTAL SALES"
 FROM authors AS au
INNER JOIN titleauthor as ta 
ON au.au_id=ta.au_id
INNER JOIN sales as s
ON ta.title_id=s.title_id
GROUP BY au.au_id,au.au_lname,au.au_fname
ORDER BY sum(s.qty) DESC LIMIT 3;""",engine)
df3
```

# Challenge 4 - Best Selling Authors Ranking

```python
df4 = pd.read_sql_query("""SELECT au.au_id as "AUTHOR ID", au.au_lname as "LAST NAME", au.au_fname as "FIRST NAME", IFNULL(sum(s.qty),0) as "TOTAL SALES"
 FROM authors AS au
INNER JOIN titleauthor as ta 
ON au.au_id=ta.au_id
INNER JOIN sales as s
ON ta.title_id=s.title_id
GROUP BY au.au_id,au.au_lname,au.au_fname
ORDER BY sum(s.qty) desc;""", engine)
df4
```
