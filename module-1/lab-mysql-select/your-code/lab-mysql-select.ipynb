```python
import pandas as pd
```

```python
df = pd.read_sql_query("""SELECT

a.au_id as "AUTHOR ID",
au_lname as "LAST NAME",
au_fname as "FIRST NAME",
t.title as "TITLE",
p.pub_name as "PUBLISHER"

from authors as a 
left join titleauthor as ti on a.au_id = ti.au_id
left join titles as t on t.title_id  = ti.title_id
left join publishers as p on t.pub_id = p.pub_id;
""",engine)
df
```

```python
df = pd.read_sql_query("""SELECT

a.au_id as "AUTHOR ID",
au_lname as "LAST NAME",
au_fname as "FIRST NAME",
p.pub_name as "PUBLISHER",
count(t.title) as "TITLE COUNT"

from authors as a 
left join titleauthor as ti on a.au_id = ti.au_id
left join titles as t on t.title_id  = ti.title_id
left join publishers as p on t.pub_id = p.pub_id

group by 1,2,3,4;""", engine)

df
```

```python
df = pd.read_sql_query("""select 
a.au_id as "AUTHOR ID",
au_lname AS "LAST NAME",
au_fname AS "FIRST NAME",
SUM(qty) as "TOTAL"

from authors as a 
left join titleauthor  as ta on a.au_id = ta.au_id
left join sales as s on ta.title_id = s.title_id


group by 1,2,3 
order by TOTAL desc;""",engine)
display(df.head(3))
```
