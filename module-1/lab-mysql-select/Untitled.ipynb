```python
df=pd.read_sql_query("""
SELECT authors.au_id  `AUTHOR ID`,  au_lname   `LAST NAME`, au_fname `FIRST NAME`, sum(qty) `TOTAL QUANTITY` FROM authors
    LEFT JOIN  titleauthor ON authors.au_id=titleauthor.au_id
    LEFT JOIN titles ON titleauthor.title_id=titles.title_id
    LEFT JOIN publishers ON titles.pub_id=publishers.pub_id
    LEFT JOIN sales ON titles.title_id=sales.title_id
    GROUP BY authors.au_id, au_lname, au_fname
    ORDER BY `TOTAL QUANTITY` DESC
    LIMIT 3
""", engine)
df
```

```python
df=pd.read_sql_query("""SELECT `AUTHOR ID`,    `LAST NAME`, `FIRST NAME`,  TITLES, IFNULL(TOTAL, 0) TOTAL  FROM (
SELECT authors.au_id  `AUTHOR ID`,  au_lname   `LAST NAME`, au_fname `FIRST NAME`, count(title) TITLES, sum(qty)  TOTAL  FROM authors
    LEFT JOIN  titleauthor ON authors.au_id=titleauthor.au_id
    LEFT JOIN titles ON titleauthor.title_id=titles.title_id
    LEFT JOIN publishers ON titles.pub_id=publishers.pub_id
    LEFT JOIN sales ON titles.title_id=sales.title_id
    GROUP BY authors.au_id, au_lname, au_fname
    ORDER BY `TOTAL` DESC) AS  NOTNULL""",engine)
df
```
