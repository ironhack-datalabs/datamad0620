```python
import sqlalchemy as db
from getpass import getpass
import pandas as pd
```

```python
password = getpass ("Mysql Password: ")
```

```python
dbname = "Books"
connectionData = f"mysql+pymysql://root:{password}@localhost/{dbname}"
```

```python
engine = db.create_engine(connectionData)
print("Connected to server!")
```

```python
df = pd.read_sql_query ("""SELECT titles.title_id AS 'Title ID', titleauthor.au_id AS 'AuthorID' , (titles.price * sales.qty * titles.royalty / 100 * titleauthor.royaltyper / 100) AS 'sales_royalty' 

FROM titleauthor
INNER JOIN titles
	ON titleauthor.title_id = titles.title_id
INNER JOIN sales
	ON sales.title_id = titles.title_id
    """ , engine)
df
```
